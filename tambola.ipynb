```python
import random
import numpy as np
from time import sleep
i=1
l=[]
done=[]
s=''
for i in range(1,91):
    if i%10!=0:
        l.append(str(i)+',')
    else:
        if(i!=90):
            l.append(str(i)+';')
        else:
            l.append(str(i))
    i+=1
for k in l:
    s +=k
ss= "'"+ s + "'"
y = np.matrix(s)
print(y,"\n\n")
sleep(5)

def numb():
    
    def printout(j):
        q=j-1
        m=int(q/10)
        n=q%10
        y[m,n]=0
        print("\n\n",y)
        
        
    def val():
        j = random.randint(1,90)
        
        if j not in done:
            done.append(j)
            print("New Number: ",j)
            printout(j)
            print("\nRecent:",done)
        else:
            val()
     
    
        
    
    def choice():
        if len(done)!=90:
            ask = input("\nPress Y to continue and N to quit\n")
            if ask=='Y' or ask=='y':
                print("\n\n")
                numb()
            elif ask=='N' or ask=='n':
                print("Thanks for playing")
            else:
                choice()
        else:
            print("Game over")
        
    
    

        
        
    val()
    choice()
    
numb()
```
